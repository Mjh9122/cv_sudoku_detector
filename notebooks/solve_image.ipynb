```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import puzzle_finder as pf
import solver
```

```python
# Import puzzle image and find sudoku puzzle

im = cv2.imread('../sudoku_pics/sudoku002.jpg')
puzzle = pf.find_puzzle(im)
plt.imshow(puzzle)
```

```python
nums_only = pf.isolate_nums(puzzle)
plt.imshow(nums_only)
```

```python
squares = pf.split_squares(nums_only)
number_mask = pf.mask_numbers(squares)
pf.center_squares(squares, number_mask)
```

```python
board = pf.apply_model(squares, number_mask)
print(board)
```

```python
solved, iters = solver.solve_sudoku(board)
print(f'Solved in {iters} iterations')
print(board)
```
